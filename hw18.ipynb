```python
class Account:
    def __init__(self, balance: float, acc_number: int):
        self.balance = balance
        self.acc_number = acc_number
    
    def balance_info(self):
        return self.balance


class SavingsAccount(Account):
    def __init__(self, balance: float, acc_number: int, percent: float):
        super().__init__(balance, acc_number)
        self.percent = percent

    def interest(self):
        self.balance += self.balance*(self.percent/100)

        
class CurrentAccount(Account):
    def __init__(self, balance: float, acc_number: int, odraft_lim : float):
        super().__init__(balance, acc_number)
        self.odraft_lim = odraft_lim

    def is_odraft(self):
        if self.balance < 0 :
            print (f"Acc {self.acc_number} is in overdraft")


class Bank:
    def __init__(self):
        self.accounts = []

    def update_accounts(self):
        for account in self.accounts:
            if isinstance(account, SavingsAccount):
                account.interest()
            elif isinstance(account, CurrentAccount):
                account.is_odraft()

    def open_account(self, account):
        self.accounts.append(account)

    def close_account(self, account):
        self.accounts.remove(account)

    def pay_dividend(self, dividends):
        for account in self.accounts:
            account.balance += dividends


```

```python
import  unittest

class TestBankMethods(unittest.TestCase):

    def test_open_account(self):
        bank = Bank()

        standart_ac = Account(5000, 1)
        deposit_ac = SavingsAccount(25000, 2, 25)
        credit_ac = CurrentAccount(-500, 3, 2000)

        bank.open_account(standart_ac)
        bank.open_account(deposit_ac)
        bank.open_account(credit_ac)

        self.assertIn(standart_ac, bank.accounts)
        self.assertIn(deposit_ac, bank.accounts)
        self.assertIn(credit_ac, bank.accounts)

        self.assertEqual(standart_ac.balance, 5000)
        self.assertEqual(deposit_ac.balance, 25000)
        self.assertEqual(credit_ac.balance, -500)

    def test_update_accounts(self):
        bank = Bank()

        deposit_ac = SavingsAccount(25000, 2, 25)
        bank.open_account(deposit_ac)

        mock_savings_ac = SavingsAccount(10000, 4, 20)
        bank.open_account(mock_savings_ac)

        with unittest.mock.patch('sys.stdout') as mock_stdout:
            bank.update_accounts()

            # Check if the interest method was called on the mock account
            self.assertTrue(mock_savings_ac.interest_called)
            
            # Check if the expected message was printed
            mock_stdout.assert_called_with()
            mock_stdout.write.assert_called_with("Interest added for account 4")

```
