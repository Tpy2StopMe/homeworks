```python
# 3. Write a program that will simulate user scores in a game. Create a list with 5 players’ names after that simulate 100 rounds for each player. 
# As a result of the game create a list with the player's name and score (0-1000 range). 
# And save it to a CSV file. The file should look like this:

# Player name, Score
# Josh, 56
# Luke, 784
# Kate, 90
# Mark, 125
# Mary, 877
# Josh, 345
# ...

import csv

# How is better to do naming of variables in such case? Is that would be ok?

def game_simultaion(num_of_players:int, per_round_max_score: int, sim_num: int) -> csv:
    
    def final_results(num_of_players:int, per_round_max_score: int, sim_num: int) -> list:
            
        def total_score(per_round_max_score: int, sim_num: int) -> int:
            player_score = 0
            for i in range(sim_num):
                player_score += random.randint(0, per_round_max_score)
            return player_score
        
        num_of_players = int(input("Enter numbers of players"))
        final_score_list = []
        
        for i in range(num_of_players):
            players_name = input("Enter player's name")
            final_score_list.append([players_name, total_score(per_round_max_score, sim_num)])
        
        return final_score_list
        
    with open('score_board.csv', mode='w', newline="", encoding='utf-8') as file: 
        writer = csv.writer(file)
        writer.writerow(['Player name', 'Score'])
        writer.writerows(final_results(num_of_players, per_round_max_score, sim_num))


game_simultaion(5,10,100)
        
# lineterminator='\n'


# Second ver. If we need just to generate score between 0 and 1000 without sum of each round.

# def random_score(n: int) -> int:
#     for i in range(n):
#         return random.randint(0, 1000)

# players_list = ["Pavlo", "Stepan", "Mariya", "Alina", "Nazar"]
# score_list = [[elem, random_score(100)] for elem in players_list]


# with open('score_board_2.csv', mode='w', newline="", encoding='utf-8') as file: 
#     writer = csv.writer(file)
#     writer.writerow(['Player name', 'Score'])
#     writer.writerows(score_list)

# Third ver. With dict.

# def random_score(n: int) -> int:
#     for i in range(n):
#         return random.randint(0, 1000)

# players_list = ["Павло", "Степан", "Марія", "Аліна", "Назар"]
# score_list = [{"Player name": elem, "Score": random_score(100)} for elem in players_list]

# with open('score_board_3.csv', mode='w', newline="", encoding='utf-8', ) as file:
#     header = ['Player name', 'Score']
#     writer = csv.DictWriter(file, fieldnames=header)
    
#     writer.writeheader()
#     writer.writerows(score_list)
```

```python
# # Write a script that reads the data from the previous CSV file and 
# creates a new file called high_scores.csv where each row contains the player name and their highest score. 
# The final score should be sorted by descending to the highest score. 
# The output CSV file should look like this:

# Player name, Highest score
# Kate, 907
# Mary, 897
# Luke, 784
# Mark, 725
# Josh, 345

with open('score_board.csv','r', newline="") as score_board, open('leaderbord.csv','w', newline="") as leaderbord:
    reader = csv.reader(score_board)
    writer = csv.writer(leaderbord)
    
    data = []
    for row in reader:
        data.append(row)
    data = sorted(data[1:], reverse=True, key=lambda data:data[1])
 
    writer.writerow(['Player name', 'Highest score'])
    writer.writerows(data)
```
